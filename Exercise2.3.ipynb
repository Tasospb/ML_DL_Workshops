```python
d = f.loc[:,["Total","HP","Attack"]]
d.median()
```

```python
import matplotlib.pyplot as plt

# Calculate means and variances
means = d.mean()
variances = d.var()

# Create bar chart
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Plot means
means.plot(kind='bar', ax=ax1, color='skyblue')
ax1.set_title('Average Values')
ax1.set_ylabel('Value')
ax1.grid(True, alpha=0.3)

# Plot variances
variances.plot(kind='bar', ax=ax2, color='lightgreen')
ax2.set_title('Variances')
ax2.set_ylabel('Value')
ax2.grid(True, alpha=0.3)

# Adjust layout to prevent overlap
plt.tight_layout()

# Add value labels on top of each bar
for ax in [ax1, ax2]:
    for i, v in enumerate(ax.containers[0]):
        ax.text(v.get_x() + v.get_width()/2, 
                v.get_height(), 
                f'{v.get_height():.2f}', 
                ha='center', 
                va='bottom')

plt.show()

# Print numerical values
print("\nMean values:")
print(means)
print("\nVariances:")
print(variances)
```
